```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv("housingprice.csv")
data
```

```python
%matplotlib inline
plt.xlabel('area(aqr ft)')
plt.ylabel('price(INR)')
plt.scatter(data.area,data.price, color='green', marker='*')
```

```python
model = linear_model.LinearRegression()
model.fit(data[['area']],data.price)
```

```python
model.predict([[3300]])
```

```python
%matplotlib inline
plt.xlabel('area(aqr ft)')
plt.ylabel('price(INR)')
plt.scatter(data.area,data.price, color='green', marker='*')
plt.plot(data.area,model.predict(data[['area']]),color='red')
```

```python
df = pd.read_csv("areaQ.csv")
```

```python
p = model.predict(df)
```

```python
df['predictedprices'] = p
df.to_csv("NewPrediction.csv",index=False)
```
